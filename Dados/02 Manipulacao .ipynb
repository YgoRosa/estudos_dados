Exercícios de visualização

```python
import pandas as pd
import requests
```

```python
df = pd.read_csv("pokemon.csv")
```

```python
# Imprimir atributos dos tipo fogo
fire_type = df['Type 1'] == 'Fire'
print(df[fire_type][["Name", "Attack", "Defense"]])
```

```python
# Imprimir atributos dos tipo água
water_type = df["Type 1"] == "Water"
print(df[water_type][["Name","Attack","Speed"]])
```

```python
# Imprimir atributos dos tipo água e ataque maior que 80
strong_water_type = df[(df["Type 1"] == "Water") & (df["Attack"] > 80)]
print(strong_water_type[["Name","Attack","Speed"]])
```

```python
# Pegar as linhas 10 até 20, exibindo atributos específicos

pokemon_10_a_20 = df.iloc[9:20]
print(pokemon_10_a_20[["Name","Type 1","HP"]])
```

```python
# Média de ataque e defesa por tipo

attack_type_media = df.groupby("Type 1")["Attack"].mean()
print(attack_type_media)
```

```python
# Média de defesa e defesa por tipo

defense_type_media = df.groupby("Type 1")["Defense"].mean()
print(defense_type_media)
```

```python
# Contando quantos pokemons tem por região

quantity_by_region = df.groupby("Generation")["Name"].count()
print(quantity_by_region)
```

```python
# Quantos pokemons existem em cada combinação
pokemon_type_by_region = df.groupby(["Generation", "Type 1"])["Name"].count()
print(pokemon_type_by_region)
```

```python
# Ordenar por Ataque

df_order = df.sort_values("Attack", ascending=True)
print(df_order[["Name","Attack"]])
```

```python
# Pokémon do tipo Grass com HP > 60, ordenados por Defense (decrescente).

grass_type = df[((df["Type 1"] == "Grass") | (df["Type 2"] == "Grass")) & (df["HP"] > 60)]
grass_type = grass_type.sort_values("HP")
print(grass_type[["Name","HP"]])
```

```python
# Nova coluna e mostrando os 5 maiores

df["TotalStats"] = df["Attack"] + df["Defense"] + df["HP"]

df_sort_total_stats = df.sort_values("TotalStats", ascending=False)
df_sort_total_stats.head() 
```

```python
# duplicados = df[df.duplicated("Generation")]
# print(duplicados)

df_sem_dup = df.drop_duplicates("Generation")
print(df_sem_dup)
```

```python
# DESAFIO - Mostrar uma tabela com Defesa e Ataques médios por Tipo e contar quantos de cada tem

tabela = (
    df.groupby("Type 1")
      .agg({
          "Attack": "mean",
          "Defense": "mean",
          "Name": "count"
      })
      .rename(columns={"Name": "Count"})
      .sort_values("Attack")
)

print(tabela)
```
