Carregando os dados

```python
import json
import pandas as pd 
import logging
import requests

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

logging.info("Iniciando...")

url = "https://pokeapi.co/api/v2/pokemon"

logging.info(f"Chamando API: {url}")

data = None

try: 
    r = requests.get(url)
    r.raise_for_status()
    data = r.json()
    logging.info("JSON recebido com sucesso!")

except requests.exceptions.HTTPError:
    logging.error("Erro HTTP! URL não encontrada")

except requests.exceptions.RequestException:
    logging.error("Erro na conexão")
```

Mostrando status de cada pokemon

```python
for p in data['results']:
    pokemon_url = p['url']
    resp = requests.get(pokemon_url).json()
    print(p['name'].upper())  
    for i, s in enumerate(resp["stats"]):
        print(i, s["stat"]["name"], s["base_stat"])
    print()
```
